```python
# Dependencies
import pandas as pd 
```

```python
# load CSV
# File to Load (Remember to Change These)
file_to_load = "Resources/Survey-Data2.csv"
#file_to_load = "Resources/2016-FCC-New-Coders-Survey-Data.csv"

# Read Purchasing File and store into Pandas data frame
surveydata2_df = pd.read_csv(file_to_load, low_memory = False)
surveydata2_df.head()
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
newsurveydata2 = surveydata2_df.iloc[:,[0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
newsurveydata2
```

```python
# Change reading 0 and 1 to No and Yes, respectively
newsurveydata2.replace(to_replace = [0.0,1.0],value =['No','Yes'], inplace = True)
newsurveydata2.tail()
```

```python
# Extract rows for only those who attended a bootcamp
bootcamp_attendants2 = newsurveydata2.loc[(newsurveydata2['AttendedBootcamp'] =='Yes')]
bootcamp_attendants2.head()
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
large_attendance = pd.DataFrame(bootcamp_attendants2['BootcampName'].value_counts())

large_attendance.reset_index(inplace=True)
large_attendance.columns = ['BootcampName','AttendeesCount']
large_attendance.sort_values(by=['AttendeesCount'], inplace=True, ascending = False)
large_attendance.head()
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
recommended_bootcamp = bootcamp_attendants2.loc[bootcamp_attendants2['BootcampRecommend'] =='Yes','BootcampName'].value_counts()
recommended_bootcamp = pd.DataFrame(recommended_bootcamp)
recommended_bootcamp.reset_index(inplace=True)
recommended_bootcamp.columns = ['BootcampName','RecommendCount']
recommended_bootcamp
```

```python
# Merge the two created data frames on the name of tbe bootcamp
bootcampmerge = pd.merge(large_attendance, recommended_bootcamp, on='BootcampName')
bootcampmerge
```

```python
# Calculate percentage of each bootcamp's students who are recommenders
percent_recommended_students = round(bootcampmerge['RecommendCount'] / bootcampmerge['AttendeesCount'] * 100,2)
bootcampmerge['Recommend %'] = percent_recommended_students 
bootcampmerge
```
